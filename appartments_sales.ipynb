## Предобработка данных

### Добавление новых столбцов

```python
# столбец с ценой в миллионах рублей
data['last_price_mln'] = data['last_price']/1000000
```

```python
# столбцы со значениями 'airports_nearest','parks_nearest', 'ponds_nearest' и 'cityCenters_nearest' в километрах
data['airports_nearest_km'] = data['airports_nearest']/1000
data['parks_nearest_km'] = data['parks_nearest']/1000
data['ponds_nearest_km'] = data['ponds_nearest']/1000
data['cityCenters_nearest_km'] = data['cityCenters_nearest']/1000
data.head(10)
```

### Удаление дубликатов в столбце с названиями населенного пункта

```python
# исходя из списка уникальных значений столбца 'locality_name', в нем есть неявные дубликаты по типу 'Мурино', 'поселок Мурино', 'посёлок Мурино'
# для их удаления можно заменить вспомогательные слова типа "поселок" на пустую строку
data['locality_name'] = data['locality_name'].str.replace('ё', 'е', regex = True) # замена 'ё' на 'е'
# список вспомогательных слов
location_type = ['городской поселок', 'поселок', 'деревня', 'село', 'поселок городского типа', 'садоводческое некоммерческое товарищество', 'садовое товарищество', 'коттеджный поселок']
# цикл, заменяющий вспомогательные слова на пустую строку
for type_name in location_type: # перебрать список вспомогательных слов
    data['locality_name'] = data['locality_name'].replace(type_name, '', regex=True) # заменить вспомогательную строку на пустую строку
        
data.head(10)
```

**Вывод**

В столбце 'locality_name' были выявлены неявные дубликаты, которые получаются при внесении в название населенного пункта вспомогательных слов с описанием типа населенного пункта, а также применением буквы "ё".
Эти дубликаты были выявлены, буква "ё" заменена на "е", а также вспомогательные слова были заменены пустой сторокой.

###  Замена пропущенных значений

```python
# замена пропущенных значений в столбце 'balcony' на 0
data['balcony'] = data['balcony'].fillna(value=0)
data['balcony'].isnull().sum() # проверка замены
```

```python
# В стобце 'locality_name' есть пустые значения, но их количество довольно невелико от общего числа значений,
# поэтому их можно удалить
data.dropna(subset = ['locality_name'], inplace=True)
data['locality_name'].isnull().sum()
```

```python
# заполнение пропусков в столбце 'floor' на медианные значения домов в соответсвии с их локациями
data['floors_total'] = data['floor'].fillna(data.groupby('locality_name')['floors_total'].transform('median'))
data['floors_total'].isnull().sum()
```

```python
data.loc[data['floor'] > data['floors_total']].shape[0] # проверка количества строк, в которых этаж больше, чем количество этажей в доме
```

```python
list = ['airports_nearest_km', 'cityCenters_nearest_km', 'parks_around3000', 'ponds_around3000', 'ponds_nearest_km', 'parks_nearest_km']
# замена значений из столбца list на средние значения для каждой группы по локации
for i in list:
    print(i)
    print("До замены:", data[i].isnull().sum())
    data[i] = data[i].fillna(data.groupby('locality_name')[i].transform('median'))
    print("После замены:", data[i].isnull().sum())
    print()
```

```python
# заполнение оставшихся пропусков в столбцах 'airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'ponds_around3000', 'ponds_nearest', 'parks_nearest'
# с помощью маркера
list = ['airports_nearest', 'cityCenters_nearest', 'parks_around3000', 'ponds_around3000', 'ponds_nearest', 'parks_nearest']
for i in list:
    print(i) # для каждого элемента списка вывести название этого элемента
    data[i] = data[i].fillna(value=1000000) # заменить пропуск на меркер 1000000
    print(data[i].isnull().sum()) # вывести сумму пропущенных значений
```

```python
# заполнение столбца 'ceiling_height' типичным значением в зависмости от этажности здания,
# из тех соображений, что для России характерна типовая застройка, и здания с определенной этажностью 
# имеют одинаковую высоту потолка
data['ceiling_height'] = data['ceiling_height'].fillna(data.groupby('floors_total')['ceiling_height'].transform('median'))
data['ceiling_height'].isnull().sum()
```

```python
# после заполнения пропусков в 'ceiling_height' все еще осталось небольшое количество пропусков
# они будут заполнены медианой, чтобы не смазать общую статистику
data['ceiling_height'] = data['ceiling_height'].fillna(value=data['ceiling_height'].median())
data['ceiling_height'].isnull().sum()
```

```python
# заполнение пропусков в столбце 'kitchen_area'
data['total_area_round'] = data['total_area'].round(-1) # создание столбца с округленными до десятков значениями площади
data['kitchen_area'] = data['kitchen_area'].fillna(data.groupby('total_area_round')['kitchen_area'].transform('median')) # заполение пропусков медианой по каждой группе площади
data['kitchen_area'].isnull().sum()
```

```python
data['kitchen_area'] = data['kitchen_area'].fillna(value=data['kitchen_area'].median()) # замена оставшегося пропуска в 'kitchen_area' медианой по 'kitchen_area'
data['kitchen_area'].isnull().sum()
```

```python
# заполение пропусков в столбце 'living_area' 
data['living_area'] = data['living_area'].fillna(data.groupby('total_area_round')['living_area'].transform('median'))
data['living_area'].isnull().sum()
```

```python
# заполнение оставшихся пропусков в 'living_area' медианой по столбцу 'living_area'
data['living_area'] = data['living_area'].fillna(value=data['living_area'].median())
data['living_area'].isnull().sum()
```

```python
len(data.loc[data['kitchen_area'] + data['living_area'] > data['total_area']]) # выявление количества строк, где сумма площади кухни и жилой площади больше общей площади
```
